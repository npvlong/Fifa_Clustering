```python
data = data[data.overall>85]

data = data.fillna(value=0)
```

```python
data
```

```python
names = data.short_name.tolist()

data = data.drop(['short_name'], axis = 1)

data.head()
```

```python
from sklearn import preprocessing

x = data.values
scaler = preprocessing.MinMaxScaler()
x_scaled = scaler.fit_transform(x)
data_scaled = pd.DataFrame(x_scaled)
```

```python
data_scaled
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components = 2)
data2dim = pd.DataFrame(pca.fit_transform(data_scaled))
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 4)

kmeans = kmeans.fit(data2dim)

labels = kmeans.predict(data2dim)

centroid = kmeans.cluster_centers_

clusters = kmeans.labels_.tolist()
```

```python
data2dim['clusters'] = clusters
data2dim['names'] = names
data2dim.columns = ['x', 'y', 'clusters', 'names']
data2dim.head()
```

```python
sns.set(style="white")

ax = sns.lmplot(x="x", y="y", hue='clusters', data = data2dim, legend=False,
                   fit_reg=False, height = 15, scatter_kws={"s": 250})

texts = []
for x, y, s in zip(data2dim.x, data2dim.y, data2dim.names):
    texts.append(plt.text(x, y, s))

ax.set(ylim=(-2, 2))
plt.tick_params(labelsize=15)
plt.xlabel("X", fontsize = 20)
plt.ylabel("Y", fontsize = 20)

plt.show()
```
